```python
from skimage.io import imread, imshow
from skimage import io, img_as_ubyte
from skimage import measure
from skimage import filters
import matplotlib.pyplot as plt
%matplotlib inline

image_gray = imread('seg3-removebg-preview.png', as_gray=True)
imshow(image_gray)
```

```python
# 이미지를 8-bit로 변환합니다.
image_gray_8 = img_as_ubyte(image_gray)

# 이미지를 inverse 시킵니다.
inverse_image = 255 - image_gray_8

# 결과를 저장합니다.
io.imsave('inverse_image.png', inverse_image)
imshow(inverse_image)
```

```python
all_labels = measure.label(inverse_image)
blobs_labels = measure.label(inverse_image, background=0)
```

```python
plt.figure(figsize=(9, 3.5))
plt.subplot(131)
plt.imshow(inverse_image, cmap='gray')
plt.axis('off')
plt.subplot(132)
plt.imshow(all_labels, cmap='nipy_spectral')
plt.axis('off')
plt.subplot(133)
plt.imshow(blobs_labels, cmap='nipy_spectral')
plt.axis('off')

plt.tight_layout()
plt.show()
```
