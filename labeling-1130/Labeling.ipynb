```python
from skimage.io import imread, imshow
from skimage import measure
from skimage import filters
import matplotlib.pyplot as plt
%matplotlib inline

image_gray = imread('2.jpg', as_gray=True)
imshow(image_gray)
```

```python
all_labels = measure.label(image_gray)
blobs_labels = measure.label(image_gray, background=0)
```

```python
plt.figure(figsize=(9, 3.5))
plt.subplot(131)
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.subplot(132)
plt.imshow(all_labels, cmap='nipy_spectral')
plt.axis('off')
plt.subplot(133)
plt.imshow(blobs_labels, cmap='nipy_spectral')
plt.axis('off')

plt.tight_layout()
plt.show()
```
